```python
import torch
from diffusers import DiffusionPipeline

# switch to "mps" for apple devices
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.bfloat16, device_map="cuda")
```

```python
# import torch
# from diffusers import DiffusionPipeline

# # switch to "mps" for apple devices
# pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.bfloat16, device_map="cuda")

prompt = "Astronaut in a rocket, cold color palette, muted colors, detailed, 8k"
# Reduce height, width, and num_inference_steps to use less GPU memory
image = pipe(prompt, height=512, width=512,num_inference_steps=50).images[0]
```
